```python
from faster_whisper import WhisperModel
from langchain.llms import Ollama
```

```python
model_size = "base"
```

```python
model = WhisperModel(model_size, device="cpu", compute_type="int8")
```

```python
segments, info = model.transcribe("d&d.mp3", beam_size=5)
```

```python
print("Detected language '%s' with probability %f" % (info.language, info.language_probability))
```

```python
with open("transcript.txt", "w", encoding="utf-8") as f:
    for i, seg in enumerate(segments):
        f.write(f"[{seg.start:.2f}s â†’ {seg.end:.2f}s] {seg.text.strip()}\n")
```

```python
# Loading local LLM 'mistral'
llm = Ollama(model="mistral")
```

```python
# Reading transcript file generated by Whisper Model
with open("transcript.txt", "r", encoding="utf-8") as f:
    lines = [line.strip() for line in f if line.strip()]
```

```python
# Function to check if a line is story-relevant
def is_story_relevant(line):
    prompt = f"""You are an AI Dungeon Master assistant.
Decide whether the following sentence from a D&D game is **story-relevant**.
Only reply with "Yes" if it contains character actions, events, NPC interaction, lore, or quests.
Reply "No" to dice rolls, rules, jokes, or out-of-character chatter.

Line: "{line}"
Relevant?"""
    
    result = llm(prompt).strip().lower()
    return result.startswith("yes")
```

```python
# Filtering relevant lines
relevant_lines = [line for line in lines if is_story_relevant(line)]
```

```python
# Saving relevant lines into a new file
with open("filtered_transcript.txt", "w", encoding="utf-8") as f:
    for line in relevant_lines:
        f.write(line + "\n")
print(f"âœ… Filtered {len(relevant_lines)} out of {len(lines)} lines.")
```
